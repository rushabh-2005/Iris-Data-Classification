What is the simplest model we can think of ?

- In this case, if our baseline model is just randomly guessing the species of flower or guessing a single species for every data point, we would expect to have a model accuracy of 0.33 or 33%, since we have 3 different classes that are evenly balanced.

- So our model should atleast beat 33% accuracy

# Modelling - Simple manual model

```python
def single_feature_prediction(petal_length):
    """predicts the Iris species given the petal length. """
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.8:
        return 1
    else:
        return 2
```

# Logistic regression

Logistic regression is based on supervised learning model
Dependent variable is categorical and binary (0 or 1).


Definition :- Logistic regression is a statistical/machine learning method used for classification, most commonly binary classification(yes/no, spam/not spam, disease/healthy).

The core idea is :-

1. It predicts the probability that an input belongs to a particular class
   (e.g., class 1 vs class 0).
2. Instead of predicting a continuous value (like linear regression), it uses a sigmoid (logistic) function to squash outputs into the range 0 to 1.

# Model form 

1. Compute a linear  combination of inputs: z = w0 + w1x1 + w2x2 + w3x3 + .....+ wnxn
2. Pass it through the sigmoid function :
      (z) = 1
           -----
           1 + e^-z

   This gives:
       ^
       p = P(y = 1 | x) = (z)

class 1 if p >= threshold (often 0.5)
Class 0 otherwise

# Interpretation

Logistic regression models the log-odds of the probability of class 1 as a linear function of the inputs:
    log( p/1-p) = w0 + w1x1 + ..... + wnxn

# Training 

Parameters w are learned by maximizing likelihood (or equivalently, minimizing cross-entropy loss) using methods like gradient descent.

# uses

Email spam detection
Medical diagnosis (disease vs no disesase)
Credit scoring (default vs no default)
Any problem where the output is category, especially binary.

# CROSS VALIDATION

Cross-validation is a technique used to evaluate a machine learning model more reliably by training and testing it on multiple different data splits.

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
model = LogisticRegression(max_iter=200)
```

```python
accuracies = cross_val_score(model, x_train, y_train, cv=5, scoring="accuracy")
```

```python
def plot_incorrect_predictions(df_predictions, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="prediction_label", data=df_predictions, ax=axs[0])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="target_name", data=df_predictions, ax=axs[1])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="correct_prediction", data=df_predictions, ax=axs[2])
    axs[3].set_visible(False)
    plt.show()
```

```python
plot_incorrect_predictions(df_predictions, "petal length (cm)", "petal width (cm)")
```

# What is model tunning ?

Model tuning is the process of adjusting a models parameters (hyperparameters) to improve its performance on unseen data.

Model tuning = making a trained model perform better

```python
for reg_param in (1, 1.3, 1.8, 2, 2.3, 2.9, 3):
    print(reg_param)
    model = LogisticRegression(max_iter=200, C=reg_param)
    accuracies = cross_val_score(model, x_train, y_train, cv=5, scoring="accuracy")
    print(f"Accuracy: {np.mean(accuracies) * 100:.2f}%")
```
